Graph 1: Influenza Positive Tests

```r
library(ggplot2)
library(reshape2)
#Choosing file here
flu1<- read.csv(file.choose(),header=T)
# Display the data in the file
flu1
#
total<-melt(data = flu1, id.vars = "Week", measure.vars = c("Total.A", "Total.B")) 
#Display the 'total' table
total
#Extract Percent.Positive.A','Percent.Positive.B','Total...Tested','X..Positive' columns from flu1
dataframe1<-subset(flu1,select=c('Percent.Positive.A','Percent.Positive.B','Total...Tested','X..Positive'))
dataframe1
#Combine total and dataframe1 and store it in dataframe2
dataframe2<-cbind(total,dataframe1)
#Display dataframe2
dataframe2
dataframe2$Week <- factor(as.factor(dataframe2$Week))
#Plot the graph 
result<-ggplot(data=dataframe2, aes(Week)) +
  geom_bar(aes(y=value,fill=variable),width = NULL, stat="identity")+ xlab("Week") + ylab("Number of Positive specimens")+
theme(axis.text.x = element_text(angle = 45, hjust = 1))+
 ggtitle("Influenza positive tests reported to CDC by U.S clinical Laboratories,National Summary, 2018-2019 Season")+
scale_fill_manual(values=c('yellow','darkgreen'))+
    geom_line(aes(y=Percent.Positive.A*700,group=1,colour="yellow"),linetype = "dashed")+
    geom_line(aes(y=Percent.Positive.B*700,group=1,colour="darkgreen"),linetype = "dashed")+
    geom_line(aes(y=X..Positive*700,group=1,colour="black"))+
    scale_y_continuous(sec.axis = sec_axis(~./700, name = "Percent Positive"))+
    scale_color_discrete(name = "", labels = c("Percent Positive", "% Positive flu B","% Positive flu A"))+
#theme_minimal()+
    theme(plot.title = element_text(size = 8, face = "bold"))
result
```

Graph 2: Influenza Positive Tests Reported to CDC by U.S. Public Health Laboratories 2018-2019 Season(By Week)

Graph 4: Mortality

```r
library("reshape2")
library("ggplot2")
#Choose the file here
data <- read.csv(file.choose(), header=T)
#Display the data
data
colnames(data) <- c("Year", "Week","PercentDeathsIandN","Expected","Threshold","AllDeaths","PneumoniaDeaths","InfluenzaDeaths")
#Extract the columns from data
dataframe<-subset(data,select=c('Year','Week','PercentDeathsIandN', 'Expected', 'Threshold'))
#Set the plot width and height
options(repr.plot.width=14, repr.plot.height=4)
#Plot the graph
ggplot(dataframe,aes(x= Week))+
geom_line(aes(y=PercentDeathsIandN,group=1,colour="PercentsDeathIandN"),size=0.10)+
geom_line(aes(y=Expected,group=1,colour="Seasonal Baseline"),size=0.8)+  
geom_line(aes(y=Threshold,group=1,colour="Epidemic Threshold"),size=0.8)+  scale_colour_manual(values=c("black","red","blue"))+
facet_wrap(~Year, nrow = 1,strip.position="bottom")+
theme(panel.spacing = unit(-0.9, "lines"))+ xlim(0,53)+
ggtitle("Pneumonia and Influenza Mortality from the National Center for Health Statistics Mortality Surveillance System")+
theme(panel.grid.major = element_blank(), panel.grid.minor = element_blank(),
panel.background = element_blank(), axis.line = element_line(colour = "black"))+
theme(plot.title = element_text(size = 14, face = "bold"))
```

Graph 5: Pediatric Death

```r
#Choose the file here
deathweek<-read.csv(file.choose(), header=T)
#Display the data 
deathweek
deathweek$WEEK.NUMBER<-factor(deathweek$WEEK.NUMBER)
mixdeath<-melt(data = deathweek, id.vars = "WEEK.NUMBER", measure.vars = c("CURRENT.WEEK.DEATHS", "PREVIOUS.WEEK.DEATHS"))
a<-aggregate(deathweek$PREVIOUS.WEEK.DEATHS, by=list(SEASON=deathweek$SEASON), FUN=sum)
num<- as.vector(a['x'])
#Set the plot width and height
options(repr.plot.width=30, repr.plot.height=8)
#Plot the graph
ggplot(data=deathweek) + 
  geom_bar(data=mixdeath,aes(x=WEEK.NUMBER,y=value,fill=mixdeath$variable),stat="identity") + ylab("Number Of deaths")+xlab("Week of death")+
  theme(axis.text.x = element_text(angle = 90, hjust = 1))+
ggtitle("Number of Influenza-Associated Pediatric deaths by Week of Death: 2015-2016 season to present")+
  scale_fill_manual(values=c('skyblue','darkgreen'))+
  theme(plot.title = element_text(size = 30, face = "bold"))+
  annotate("text", label = paste("2015-2016 Number of deaths",num[[1]][1]), x = 14.5, y = 20, color = "black")+
  annotate("text", label = paste("2016-2017 Number of deaths",num[[1]][2]), x = 74.5, y = 20, color = "black")+
  annotate("text", label = paste("2017-2018 Number of deaths",num[[1]][3]), x = 128, y = 20, color = "black")+
  annotate("text", label = paste("2018-2019 Number of deaths",num[[1]][4]), x = 170, y = 20, color = "black")
```

Graph 7: Flu heat map of USA

```r
library(ggmap)
library(maptools)
#Choose the file here
heatmap<-read.csv(file.choose(),header=T)
heatmapsub<-subset(heatmap , select=c("STATENAME","ACTIVITY.LEVEL","ACTIVITY.LEVEL.LABEL"))
#colnames(heatmap)
heatmap$STATENAME<-tolower(heatmap$STATENAME)
states <- map_data("state")
states
hm1<-heatmap
hm1$STATENAME<-tolower(hm1$STATENAME)
heatmapsub$region<-tolower(heatmapsub$STATENAME)
map.df<-merge(states,heatmapsub,by='region',all.heatmapsub=T)
map.df <- map.df[order(map.df$order),]
heatmap$ACTIVITY.LEVEL<-substr(heatmap$ACTIVITY.LEVEL,6,9)
map.df$ACTIVITY.LEVEL<-substr(map.df$ACTIVITY.LEVEL,6,9)
#Se the plot width and height
options(repr.plot.width=16, repr.plot.height=8)
#Plot the graph
ggplot(map.df, aes(x=long,y=lat,group=group))+
geom_polygon(aes(fill=as.character(as.numeric(ACTIVITY.LEVEL)+81)))+
  geom_path()+
  scale_fill_manual(labels=c('Low','Low','Low','Moderate','Moderate','Moderate','High','High','High'),
                    values=c("lightgreen","green","darkolivegreen1","gold","yellow","darkgoldenrod","lightblue","darkred","red"))+
ggtitle("2018-19 Influenza Season Week 8 ending Feb 23, 2019")+
theme(plot.title = element_text(size = 20, face = "bold"))
```
